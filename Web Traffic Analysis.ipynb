```python
import pandas as pd

# Sample website traffic data
data = {
    "user_id": [1, 2, 3, 4, 5, 6],
    "page": ["Home", "Home", "Product", "Checkout", "Home", "Product"],
    "session_duration": [120, 45, 300, 60, 180, 240],  # in seconds
    "next_page": ["Product", "Exit", "Checkout", "Exit", "Product", "Exit"]
}

df = pd.DataFrame(data)

# Key Metrics
page_views = df["page"].value_counts()
avg_session_duration = df["session_duration"].mean()
drop_off_points = df[df["next_page"] == "Exit"]["page"].value_counts()

# User Journey Mapping
journeys = df.groupby("user_id")[["page", "next_page"]].apply(lambda x: list(x.values))

print("Page Views:\n", page_views)
print("\nAverage Session Duration (seconds):", avg_session_duration)
print("\nDrop-off Points:\n", drop_off_points)
print("\nUser Journeys:\n", journeys)
```
